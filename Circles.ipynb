# Python for Geometry: Circles and Their Properties
## Introduction
Circles are fundamental geometric shapes with unique properties such as chords, tangents, arcs, circumference, and sector areas. In this tutorial, we will explore these concepts using Python.

# Understanding Circle Properties

- **Chord**: A line segment with both endpoints on the circle.
- **Tangent**: A line that touches the circle at only one point.
- **Arc**: A portion of the circumference.
- **Circumference**: The total boundary length of a circle, given by:

  $$ C = 2\pi r $$

- **Sector Area**: The area enclosed by two radii and an arc, calculated as:

  $$ A = \frac{\theta}{360} \times \pi r^2 $$

where $ r $ is the radius and $ \theta $ is the sector angle.


## Calculating Circumference in Python

```python
import math

def calculate_circumference(radius):
    return 2 * math.pi * radius

radius = 5
print(f"Circumference of the circle: {calculate_circumference(radius)}")
```

## Calculating Arc Length

```python
def arc_length(radius, angle):
    return (angle / 360) * 2 * math.pi * radius

radius = 7
angle = 90  # degrees

print(f"Arc length: {arc_length(radius, angle)}")
```

## Calculating Sector Area

```python
def sector_area(radius, angle):
    return (angle / 360) * math.pi * radius ** 2

radius = 6
angle = 120  # degrees

print(f"Sector area: {sector_area(radius, angle)}")
```

## Visualizing a Circle with Chords and Tangents

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 5, color='blue', fill=False)

# Chord
plt.plot([-3, 3], [4, 4], 'g-', label="Chord")

# Tangent
plt.plot([5, 6], [0, 1], 'r-', label="Tangent")

ax.add_patch(circle)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.legend()
plt.title("Circle with Chord and Tangent")
plt.grid()
plt.show()
```

## Conclusion
Circles have fascinating geometric properties that can be explored using Python. Whether computing circumference, arc lengths, sector areas, or visualizing geometric elements, Python makes understanding circles more interactive.
