# Python for Geometry: Coordinate Geometry
## Introduction
Coordinate geometry deals with geometric properties using algebra and graphs. It allows us to analyze points, lines, slopes, distances, and midpoints in the Cartesian plane. Python provides tools for performing calculations and visualizing coordinate geometry concepts.

## Graphing in Coordinate Geometry
We use the matplotlib library to plot points, lines, and other geometric figures in the coordinate plane.

## Python Code for Plotting Points

```python
import matplotlib.pyplot as plt

# Define points
x_values = [2, 5, -3]
y_values = [3, -2, 4]

plt.scatter(x_values, y_values, color='blue', marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plotting Points")
plt.grid(True)
plt.show()
```

# Slope Formula

The slope of a line measures its steepness and is given by:

$$ m = \frac{y_2 - y_1}{x_2 - x_1} $$



## Python Code for Calculating Slope

```python
def calculate_slope(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return (y2 - y1) / (x2 - x1)

p1 = (3, 4)
p2 = (7, 10)

slope = calculate_slope(p1, p2)
print(f"Slope of the line: {slope}")
```

# Distance Formula

The distance between two points $ (x_1, y_1) $ and $ (x_2, y_2) $ is given by:

$$ d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $$



## Python Code for Calculating Distance

```python
import math

def calculate_distance(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

p1 = (1, 2)
p2 = (4, 6)

distance = calculate_distance(p1, p2)
print(f"Distance between points: {distance}")
```

# Midpoint Formula

The midpoint between two points is calculated as:

$$ M = \left( \frac{x_1 + x_2}{2}, \frac{y_1 + y_2}{2} \right) $$



## Python Code for Calculating Midpoint

```python
def calculate_midpoint(p1, p2):
    x1, y1 = p1
    x2, y2 = p2
    return ((x1 + x2) / 2, (y1 + y2) / 2)

p1 = (2, 4)
p2 = (6, 8)

midpoint = calculate_midpoint(p1, p2)
print(f"Midpoint: {midpoint}")
```

## Visualizing Lines in Coordinate Geometry
Using **matplotlib**, we can plot a line connecting two points.

```python
import numpy as np

x_values = [2, 6]
y_values = [4, 8]

plt.plot(x_values, y_values, marker='o', linestyle='-', label="Line Between Points")
plt.scatter([(x_values[0] + x_values[1]) / 2], [(y_values[0] + y_values[1]) / 2], color='red', label="Midpoint")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Coordinate Geometry Visualization")
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion
Coordinate geometry helps us analyze points, lines, slopes, distances, and midpoints. Python simplifies calculations and provides powerful visualization tools for understanding spatial relationships.
