# Python for Geometry: Quadrilaterals and Polygons
## Introduction
Quadrilaterals and polygons are fundamental shapes in geometry. In this tutorial, we will explore their properties, classifications, and how to work with them using Python.

## Understanding Quadrilaterals
A quadrilateral is a four-sided polygon with the following classifications:

* **Square:** Four equal sides and four right angles.
* **Rectangle:** Opposite sides are equal with four right angles.
* **Parallelogram:** Opposite sides are equal and parallel.
* **Rhombus:** Four equal sides with opposite angles equal.
* **Trapezoid:** Only one pair of opposite sides is parallel.

## Python Code for Identifying Quadrilaterals

```python
def classify_quadrilateral(sides, angles):
    if len(sides) != 4 or len(angles) != 4:
        return "Not a quadrilateral"
    
    if len(set(sides)) == 1 and set(angles) == {90}:
        return "Square"
    elif len(set(sides)) == 2 and set(angles) == {90}:
        return "Rectangle"
    elif len(set(sides)) == 1:
        return "Rhombus"
    elif sides[0] == sides[2] and sides[1] == sides[3]:
        return "Parallelogram"
    elif angles.count(90) == 0:
        return "Trapezoid"
    else:
        return "Other Quadrilateral"

sides = [5, 5, 5, 5]
angles = [90, 90, 90, 90]
print(f"Quadrilateral Type: {classify_quadrilateral(sides, angles)}")
```

## Understanding Polygons
A polygon is a closed shape with three or more sides. The most common types include:

* **Triangle (3 sides)**
* **Quadrilateral (4 sides)**
* **Pentagon (5 sides)**
* **Hexagon (6 sides)**
* **Octagon (8 sides)**

## Python Code for Identifying Polygons

```python
def classify_polygon(sides):
    types = {3: "Triangle", 4: "Quadrilateral", 5: "Pentagon", 6: "Hexagon", 8: "Octagon"}
    return types.get(len(sides), "Unknown Polygon")

sides = [4, 4, 4, 4, 4]
print(f"Polygon Type: {classify_polygon(sides)}")
```

# Properties of Regular Polygons

A regular polygon has equal sides and equal angles. The sum of interior angles of an $ n $-sided polygon is:

$$ \text{Sum} = (n - 2) \times 180 $$

where $ n $ is the number of sides.


## Python Code for Computing Interior Angles

```python
def sum_of_interior_angles(n):
    return (n - 2) * 180

num_sides = 6
print(f"Sum of interior angles: {sum_of_interior_angles(num_sides)} degrees")
```

## Visualizing Quadrilaterals and Polygons
Using **matplotlib**, we can plot polygons to observe their properties.

```python
import matplotlib.pyplot as plt

# Define quadrilateral points
x = [0, 4, 6, 2, 0]
y = [0, 0, 3, 4, 0]

plt.plot(x, y, marker="o", linestyle="-")
plt.fill(x[:4], y[:4], color='lightblue', alpha=0.5)

plt.xlim(-1, 7)
plt.ylim(-1, 5)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadrilateral Visualization")
plt.grid(True)
plt.show()
```
