## Introduction to Geometry
Geometry is the branch of mathematics that deals with points, lines, angles, and planes. It is foundational for understanding spatial relationships, measurements, and design. Python, with its rich set of libraries, provides powerful tools for performing geometric calculations and visualizations.

## Basic Concepts
* **Point:** A location in space, defined by coordinates (x, y) in two dimensions or (x, y, z) in three dimensions.
* **Line:** A straight path that extends infinitely in both directions.
* **Angle:** The space between two intersecting lines, measured in degrees or radians.
* **Plane:** A flat, two-dimensional surface that extends infinitely.

## Working with Points in Python
In Python, we can represent points using tuples or classes.

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

p1 = Point(3, 4)
print(p1)
````

## Lines and Distance Between Points
Using the distance formula, we can compute the distance between two points.

````python
from sympy import Point, Line

p1 = Point(3, 4)
p2 = Point(7, 1)
line = Line(p1, p2)

distance = p1.distance(p2)
print(f"Distance between points: {distance}")
````

## Angle Calculations
Angles are fundamental in geometry. We can compute angles using trigonometric functions.

````python
import math

def angle_between(v1, v2):
    dot_product = sum(a*b for a, b in zip(v1, v2))
    magnitude_v1 = math.sqrt(sum(a**2 for a in v1))
    magnitude_v2 = math.sqrt(sum(b**2 for b in v2))
    angle = math.acos(dot_product / (magnitude_v1 * magnitude_v2))
    return math.degrees(angle)

v1 = (1, 0)
v2 = (0, 1)
print(f"Angle between vectors: {angle_between(v1, v2)} degrees")
````

## Visualizing Geometry with Matplotlib
We can visualize points, lines, and shapes using **matplotlib**.

````python
import matplotlib.pyplot as plt

x_values = [3, 7]
y_values = [4, 1]

plt.plot(x_values, y_values, marker='o', linestyle='-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Between Two Points')
plt.grid(True)
plt.show()
````
